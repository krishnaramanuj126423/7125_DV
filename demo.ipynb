```python
import pandas as pd
import matplotlib.pyplot as plt
```

Bar chart


```python
name = ['Nikunj','Jugal','Harshit','Jenish','Parth']
cgpa = [8,7,5,9,6]

plt.xlabel(name)
plt.ylabel(cgpa)

plt.bar(name,cgpa)


```

Scatterplot

```python
sales = [25000,14000,18000,16000,12000]
profit = [4500,3000,3500,3200,3900]

plt.scatter(sales,profit)
plt.xlabel('Sales')
plt.ylabel('Profit')
```

line plot

```python
year = [2016,2017,2018,2019,2020]
profit = [80000,110000,90000,85000,120000]

plt.plot(year,profit)
plt.xlabel('Year')
plt.ylabel('Profit')
```

multiple line plot

```python
year = [2016,2017,2018,2019,2020]
profit = [80000,110000,90000,85000,120000]
sales  = [800000,1200000,9000000,2500000,10000000]

plt.plot(year,profit)
plt.plot(year,sales)

```
