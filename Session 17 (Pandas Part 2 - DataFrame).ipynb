# Selecting rows from a dataframe

1) `loc()`: Searches using index label (When String index)
2) `iloc()`: Searches using index position (When numeric index)

```python
students.set_index("Name", inplace = True)
```

```python
students
```

```python
# loc(); Searches using index labels

# Fetch single row
students.loc['Jenish']
```

```python
# Fetch multiple rows

students.loc[['Harshit', 'Jenish']]
```

```python
students.loc['Jugal': 'Jenish']   # in loc() 'Jenish' is included
```

```python
# Pandas has its own hidden index starting from 0, over and above our custom index.
# So when there is label index, we can also use iloc() as well

students.iloc[1:4]
```

```python
# iloc(): Searches using index position.

students.reset_index(inplace = True)

students
```

# There are always two indices for rows and columns. One which we provide and the other which Python provides internally.
# The one which we provide may be a string or integer but which python provides internally is always integer.

`For eg`: In students dataframe, `Name`, `iq`, `cgpa`, `package` are the column names which we provided, but they also have their own internal indices as 0, 1, 2, 3 respectively.

Same way in `movies` dataframe, the row indices which we are seeing is the one which we have provided(Check in the original file). But python also handles its own indices internally.

* So when our indices is are integer, we can access it using **iloc() and loc()** both and
* When our given indices are string, to fetch using those given lables, use **loc()** and to fetch using internal indices, use **iloc()**

# Filtering a dataframe.

```python
ipl.head()
```

```python
# Find all the final winners of each season

ipl['MatchNumber'] == 'Final'    # Returns boolean indexing. Now see the next step
```

```python
ipl[ipl['MatchNumber'] == 'Final']    # Returns the dataframe 
```

```python
ipl[ipl['MatchNumber'] == 'Final']    [['Season', 'WinningTeam']]

# See, we got our answer
```

```python
# How many super over occurred.

ipl[ipl['SuperOver'] == 'Y']  [['City', 'SuperOver']]
```

```python
ipl[ipl['SuperOver'] == 'Y']  [['City', 'SuperOver']].shape

# Shape shows rows and columns. So 14 times super over is played till now.

# shape[0] gives 14
```

```python
# How many matches did CSK won in Kolkata?
# Here we'll have to fetch all matches played at kolkata, and the winning team is CSK

ipl['City'] == 'Kolkata'    # Will return a boolean series of matches played at Kolkata. Now we'll put this inside our dataframe
```

```python
ipl[ipl['City'] == 'Kolkata']    # Will return a dataframe
```

```python
# Now apply another condition i.e
# Winning team is Kolkata

ipl[
(ipl['City'] == 'Kolkata') &
(ipl['WinningTeam'] == 'Chennai Super Kings')
]

# We got our final result
```

```python
# Toss winner is the match winner (In percentage)

ipl[ipl['TossWinner'] == ipl['WinningTeam']]
```

```python
# In percentage

(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]) / ipl.shape[0]
```

```python
# Action movies with rating > 7.5

# Now understand this very carefully.
# If u see in 'genres' column, you will notice that for some movies there are more that 1 genres. So we'll have to split it
# and then filter if atleast one of them is 'Action'

movies['genres'].str.split('|')
```

```python
action_movies = movies[movies['genres'].str.split('|').apply(lambda x: 'Action' in x)]

action_movies

# A dataframe of previous boolean series.
```

```python
# Action movies with rating > 7.5

action_movies[action_movies['imdb_rating'] > 7.5]
```

```python
# We can also try this as well

rating = movies['imdb_rating'] > 7.5
action_bool = movies['genres'].str.split('|').apply(lambda x: 'Action' in x)

rating
```

```python
movies[action_bool & rating] 
```

```python
# Best/ simplest way is to use contains() method

movies[
(movies['genres'].str.contains('Action')) &
(movies['imdb_rating'] > 7.5)
]
```

# Some important DataFrame's functions.

```python
# astype(): Changes the data type of a given column.

ipl.info()

# Let's change data type of 'ID' from int64 to int32
# Moreover, here the memory usage is 148.6. 
# We will again check it after changing the dtype.
```

```python
ipl['ID'] = ipl['ID'].astype('int32')

ipl.info()

# See, dtype changed and memory usage also reduced from 148 to 144.
```

```python
# We have a column 'Season' which stores the year of the match.
# We know these years will be repeating in the dataset. Hence they are called as 'Categorical column'.
# In pandas, we have a data type 'category' for catrgorical columns.
# So let's change this also

ipl['Season'] = ipl['Season'].astype('category')

ipl.info()

# See, dtype of 'Season' changed and also memory usage reduced to 138.

# lly we can also change Team1 and Team2 as 'category' bcz teams are also repeating. So they are also categorical columns.
# Let's do it.
```

```python
ipl = ipl.astype(
    {'Team1': 'category',
    'Team2': 'category'
    }
)
```
