```python
import matplotlib.pyplot as plt

categories = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [350, 480, 290, 610, 420]

plt.figure(figsize=(9, 6))

bars = plt.bar(categories, sales, color='#3498db', edgecolor='navy', alpha=0.85, width=0.6)

plt.title('Product Sales Performance', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Products', fontsize=12, labelpad=10)
plt.ylabel('Units Sold', fontsize=12, labelpad=10)

plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 10, f'{yval}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()  # Ensures everything fits perfectly without clipping
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

quarters = ['Q2', 'Q1', 'Q3', 'Q4'] # Sorted by total sales
prod_a = [35, 20, 30, 35]
prod_b = [32, 25, 34, 20]
prod_c = [20, 45, 26, 41]

prod_a = np.array(prod_a)
prod_b = np.array(prod_b)
prod_c = np.array(prod_c)

fig, ax = plt.subplots(figsize=(9, 6))

bars1 = ax.bar(quarters, prod_a, label='Product A', color='#2ecc71', width=0.5)
bars2 = ax.bar(quarters, prod_b, bottom=prod_a, label='Product B', color='#3498db', width=0.5)
bars3 = ax.bar(quarters, prod_c, bottom=prod_a+prod_b, label='Product C', color='#e74c3c', width=0.5)

ax.set_title('Quarterly Sales Breakdown by Product', fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('Quarters', fontsize=12, labelpad=10)
ax.set_ylabel('Units Sold', fontsize=12, labelpad=10)
ax.grid(axis='y', linestyle='--', alpha=0.5)
ax.legend(loc='upper left')

for i in range(len(quarters)):
    ax.text(i, prod_a[i]/2, str(prod_a[i]), ha='center', va='center', color='white', fontweight='bold')
    ax.text(i, prod_a[i] + prod_b[i]/2, str(prod_b[i]), ha='center', va='center', color='white', fontweight='bold')
    ax.text(i, prod_a[i] + prod_b[i] + prod_c[i]/2, str(prod_c[i]), ha='center', va='center', color='white', fontweight='bold')
    total = prod_a[i] + prod_b[i] + prod_c[i]
    ax.text(i, total + 2, str(total), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.savefig('stacked_bar_plot.png', dpi=300)
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)  # Setting a seed ensures the "random" data stays identical every run
exam_scores = np.random.normal(loc=75, scale=10, size=250)

fig, ax = plt.subplots(figsize=(9, 6))

n, bins, patches = ax.hist(
    exam_scores, 
    bins=15, 
    color='#34495e',       # Charcoal gray fill
    edgecolor='white',     # Crisp white borders between bins
    alpha=0.85
)

ax.set_title('Distribution of Student Exam Scores', fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('Exam Scores (Percentage)', fontsize=12, labelpad=10)
ax.set_ylabel('Number of Students (Frequency)', fontsize=12, labelpad=10)

ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)

class_a = np.random.normal(70, 10, 100)
class_b = np.random.normal(75, 12, 100)
class_c = np.random.normal(65, 8, 100)

data = [class_a, class_b, class_c]
labels = ['Class A', 'Class B', 'Class C']

fig, ax = plt.subplots(figsize=(9, 6))
box = ax.boxplot(
    data,
    patch_artist=True,
    notch=False,
    flierprops=dict(marker='o', markerfacecolor='#e74c3c', markersize=6, linestyle='none', markeredgecolor='none')
)
ax.set_xticks(range(1, len(labels) + 1))
ax.set_xticklabels(labels)

colors = ['#3498db', '#2ecc71', '#9b59b6']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
    patch.set_edgecolor('#2c3e50')
    patch.set_linewidth(1.5)


for whisker in box['whiskers']:
    whisker.set(color='#2c3e50', linewidth=1.5, linestyle='--')

for cap in box['caps']:
    cap.set(color='#2c3e50', linewidth=1.5)

for median in box['medians']:
    median.set(color='#e74c3c', linewidth=2)


ax.set_title('Exam Score Distribution Across Classes', fontsize=16, fontweight='bold', pad=15)
ax.set_xlabel('Classes', fontsize=12, labelpad=10)
ax.set_ylabel('Scores', fontsize=12, labelpad=10)

ax.grid(axis='y', linestyle=':', alpha=0.6)

plt.tight_layout()
plt.savefig('box_plot.png', dpi=300)
```

```python
import matplotlib.pyplot as plt
import numpy as np

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig, ax = plt.subplots(figsize=(10, 7), subplot_kw={"projection": "3d"})

surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.9)

ax.set_title('3D Surface Plot', fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('X Axis', fontsize=12, labelpad=10)
ax.set_ylabel('Y Axis', fontsize=12, labelpad=10)
ax.set_zlabel('Z Axis', fontsize=12, labelpad=10)

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, pad=0.1)

plt.tight_layout()
plt.savefig('3d_surface_plot.png', dpi=300)
```
