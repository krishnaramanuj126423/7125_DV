Bar Chart 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
name = ['Nikunj','Jugal','Harshit','Jenish','Parth']
cgpa = [8,7,5,9,6]

plt.xlabel(name)
plt.ylabel(cgpa)

plt.bar(name,cgpa)
```

Scatter Plot

```python
sales = [25000,14000,16000,18000,12000]
profit = [4500,3500,3000,3200,3900]

plt.scatter(sales, profit)
plt.xlabel('sales')
plt.ylabel('profit')
```

Line Plot

```python
year = [2016,2017,2018,2019,2020]
profit = [4500,3500,3000,3200,3900]

plt.plot(year, profit)
plt.xlabel('year')
plt.ylabel('profit')
```

Multiple Line Plot

```python
year = [2016,2017,2018,2019,2020]
profit = [80000,110000,90000,85000,120000]
sales = [800000,1200000,900000,2500000,1200000]

plt.plot(year,profit, color='red', label = 'Year VS Profit')
plt.plot(year,sales, color='green', label = 'Year VS Sales')

plt.xlabel('year')
plt.ylabel('sales/profit')

plt.legend()
```

Histogram

```python
marks = [50,54,59,65,68,70,73,75,81,83,88,90]

plt.hist(marks,bins = 4)
```

BoxPlot

```python
marks = [50,54,59,65,68,70,73,75,81,83,88,90,91,92,93,94,95,96,97,98,99,100]
plt.boxplot(marks)
```

Pie Chart

```python
name = ['Nikunj','Jugal','Harshit','Jenish','Parth']
cgpa = [8,7,5,9,6]
explode = [0,0,0.1,0,0]

plt.pie(x = cgpa, labels = name, autopct = "%f0.1%%", explode = explode)
```

```python
x = [1,4,7,10]
y = [2,5,8,11]
z = [3,6,9,12]

fig, axis = plt.subplots(figsize = (6,6), subplot_kw = {'projection' : '3d'})

axis.scatter3D(x,y,z)
```
