```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel("Season_23_dataset_1.xlsx")

# Ensure the PM2.5 column is numeric (handles any non-numeric coercions)
df['PM2.5'] = pd.to_numeric(df['PM2.5'], errors='coerce')

# Filter the data for 'Iran' and 'China'
df_filtered = df[df['Country'].isin(['Iran', 'China'])]

# Group by 'Year' and 'Country', and calculate the sum of 'PM2.5'
df_grouped = df_filtered.groupby(['Year', 'Country'])['PM2.5'].sum().reset_index()

# Pivot the data to have 'Year' as index and 'Country' as columns
df_pivot = df_grouped.pivot(index='Year', columns='Country', values='PM2.5')

# Create the line plot
plt.figure(figsize=(10, 6))
df_pivot.plot(kind='line', marker='o', ax=plt.gca())

# Add labels and title
plt.title('Total PM2.5 over Time in Iran and China')
plt.xlabel('Year')
plt.ylabel('Sum of PM2.5')
plt.grid(True)
plt.tight_layout()

# Save the plot to a file
plt.savefig('pm25_lineplot.png')
```

Histogram


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel("Season_23_dataset_1.xlsx")

# Convert PM10 to numeric, ignoring errors and dropping missing values
pm10_data = pd.to_numeric(df['PM10'], errors='coerce').dropna()

# Create the figure
plt.figure(figsize=(10, 6))

# Plot histogram with density=True for probability on the y-axis
plt.hist(pm10_data, bins=30, density=True, edgecolor='black', alpha=0.7)

# Add labels, title, and grid
plt.title('Histogram of PM10 (Probability)')
plt.xlabel('PM10')
plt.ylabel('Probability Density')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()

# Save the plot
plt.savefig('pm10_histogram.png')
```

Scatter Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel("Season_23_dataset_1.xlsx")

# Ensure PM2.5 and PM10 columns are numeric (handles any non-numeric strings)
df['PM2.5'] = pd.to_numeric(df['PM2.5'], errors='coerce')
df['PM10'] = pd.to_numeric(df['PM10'], errors='coerce')

# Filter for 'Poland' and 'Chile', and drop rows with missing PM2.5 or PM10 values
df_filtered = df[df['Country'].isin(['Poland', 'Chile'])].dropna(subset=['PM2.5', 'PM10'])

# Create the figure
plt.figure(figsize=(10, 6))

# Define colors for distinction
colors = {'Poland': 'blue', 'Chile': 'red'}

# Plot scatter points for each country separately to create a legend
for country in ['Poland', 'Chile']:
    subset = df_filtered[df_filtered['Country'] == country]
    plt.scatter(subset['PM2.5'], subset['PM10'], label=country, color=colors[country], alpha=0.7, edgecolors='k')

# Add labels, title, and legend
plt.title('Scatter Plot of PM2.5 vs PM10 for Poland and Chile')
plt.xlabel('PM2.5')
plt.ylabel('PM10')
plt.legend(title="Country")
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Save the plot to a file
plt.savefig('pm25_pm10_scatter.png')
```

Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel("Season_23_dataset_1.xlsx")

# Count the frequencies of each country and get the top 5
top_5_countries = df['Country'].value_counts().head(5)

# Create the figure
plt.figure(figsize=(8, 8))

# Plot the pie chart
plt.pie(top_5_countries, labels=top_5_countries.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# Add title
plt.title('Top 5 Most Frequent Countries')
plt.tight_layout()

# Save the plot to a file
plt.savefig('top5_countries_pie.png')
```

Bar Chart


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel("Season_23_dataset_1.xlsx")

# Count the frequencies of each country and get the top 5
top_5_countries = df['Country'].value_counts().head(5)

# Create the figure
plt.figure(figsize=(10, 6))

# Plot the bar chart
top_5_countries.plot(kind='bar', color='skyblue', edgecolor='black')

# Add labels, title, and formatting
plt.title('Top 5 Most Frequent Countries')
plt.xlabel('Country')
plt.ylabel('Frequency Count')

# Rotate x-axis labels to ensure they are not truncated and remain readable
plt.xticks(rotation=45, ha='right')

# Add a grid for better readability of counts
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the plot to a file
plt.savefig('top5_countries_bar.png')
```
